```python
import pandas as pd
import numpy as np
from keras.layers import Input, Dense, LSTM, RepeatVector, TimeDistributed, RNN, Lambda, Concatenate, Reshape, Add
from keras.layers.merge import Multiply
from keras.models import Model
from keras.losses import mean_squared_error, binary_crossentropy
from keras.models import Model
from keras.models import load_model
from keras.models import model_from_json
import keras.backend as K
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
from copy import deepcopy
import scipy.ndimage
def open_and_load(string):
    with open('models/'+string+'.json', 'r') as f:
        model = model_from_json(f.read())
    model.load_weights('models/'+string+'.h5')   
    return model
def save_model(model, model_str):
    model_json = model.to_json()
    with open('models/'+model_str+".json", "w") as json_file:
        json_file.write(model_json)
    # serialize weights to HDF5
    model.save_weights('models/'+model_str+".h5")
    print("Saved model to disk")
def plot_loss(model, gen):
    original_weights = deepcopy(model.get_weights())
    
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set(xlim=(-1, 1))
    ax.set_xlabel('alpha')
    ax.set_ylabel('loss')
    for _ in range(5):
        A,B = next(gen(batch_size=1))
        model.set_weights(original_weights)
        delta = deepcopy(model.get_weights())
        theta = []
        for weights in delta:
            theta += [weights * np.random.normal(loc=1.0, scale=0.5, size=weights.shape)]

        def get_weight_combo(alpha):
            l = []

            for i in range(len(delta)):
                l += [ delta[i]*(1-alpha) + theta[i]*alpha]
            return l


        rmse_list = []
        alphas = np.linspace(-1, 1, num=101)
        for alpha in alphas:
            weights = get_weight_combo(alpha)
            model.set_weights(weights)
            P = model.predict(A)
            rmse_list += [mse(B, P)**0.5]


        plt.plot(alphas, rmse_list)
    plt.show()
    model.set_weights(original_weights)
```

```python
def plot_landscape(model, gen):
    original_weights = deepcopy(model.get_weights())
    def get_random_direction_vector_like(weights):
        vector = []
        for layer in weights:
            vector += [np.random.uniform(size=layer.shape)] #* layer maybe
        assert len(vector) == len(weights)
        return vector
    def get_loss(a, b):
        W = []
        for w, d, e in zip(original_weights, delta, eta):
            assert w.shape == d.shape
            assert d.shape == e.shape
            W += [w + a * d + b * e]
        assert len(W) == len(original_weights)
        model.set_weights(W)
        X, Y = next(gen(batch_size=1))
        P = model.predict(X)
        model.set_weights(original_weights)
        return mse(P, Y)
    
    delta = get_random_direction_vector_like(original_weights)
    eta = get_random_direction_vector_like(original_weights)
    alphas = np.linspace(-0.1, 0.1, 100)
    betas = np.linspace(-0.1, 0.1, 100)
    
    con = []
    i = 0
    for alpha in alphas:
        for beta in betas:
            con += [get_loss(alpha, beta)]
            i += 1
            if i % 500 == 0: print(i,'/',len(alphas)*len(betas))
    con = np.array(con).reshape(100,100)
    print("Min: ", np.min(con))
    print("Max: ", np.max(con))
    print("Stdev: ", np.std(con))
    con = np.round(con, decimals=5)
    con = np.minimum(con, 0.2)
    
    plt.figure(figsize=(7,7))
    X, Y = np.meshgrid(alphas, betas)
    plt.contourf(X, Y, con, 10, alpha=.25, cmap=plt.cm.hot)
    C = plt.contour(X, Y, con, 10, colors='black', linewidths=0.2)
    plt.clabel(C, inline=1, fontsize=15)
    plt.xticks(())
    plt.yticks(())

    plt.show()
    return con
```

# Select border

```python
def border_att_gen(batch_size=128):
    while True:
        a = np.zeros((batch_size,3) )
        b1 = np.random.uniform(low=-10,high=0,size=batch_size)
        b2 = np.random.uniform(low=0,high=10,size=batch_size)
        x = np.random.uniform(low=b1, high=b2)
        a[:,0] = b1
        a[:,1] = b2
        a[:,2] = x
        
        isb1 = ((x-b1)/(b2-b1) < 0.5)
        
        yield a, np.array([isb1,(1-isb1)]).T
```

```python
# save_model(border_att, 'border_att')
border_att = open_and_load('border_att')
```

```python
next(border_att_gen(batch_size=1))
```

```python
i = Input(shape=(3,))
h = Dense(10, activation='relu')(i) #mby act=none
h = Dense(1)(h)
h = change(h)
h = Lambda(lambda s: K.concatenate([1-s,s]), output_shape=(2,))(h)
border_att = Model(inputs=i, outputs=h)
# border_att.compile(loss=mean_squared_error,
border_att.compile(loss=binary_crossentropy,
              optimizer='adam',
              metrics=['accuracy'])
border_att.summary()
```

```python
border_att.fit_generator(border_att_gen(), steps_per_epoch=12800, epochs=10)
```

```python
while True:
    X = next(border_att_gen(batch_size=1))
    Y = border_att.predict(X[0])
    if np.any(X[1]!=Y):
        print(X)
        print(Y)
        break
```

```python
i = Input(shape=(3,))
border_att.name = 'attention'
h = border_att(i)
border_att.trainable = False

b1b2 = Lambda(lambda s: s[:,:2], output_shape=(2,), name='b1b2')(i)
mul = Multiply()([h, b1b2]) 

out = Lambda(lambda s: s[:,0]+s[:,1], output_shape=(1,))(mul)
# out = Dense(1, use_bias=False)(mul)
border = Model(inputs=i, outputs=out)
border.compile(loss=mean_squared_error,
              optimizer='adam',
              metrics=['accuracy'])
border.summary()
```

```python
# save_model(border, 'border')
border = open_and_load('border')
```

```python
X = next(border_att_gen(batch_size=1))[0]
print(X)
border.predict(X)
```

### Move

```python
def move_gen(batch_size=128):
    while True:
        M = np.zeros((batch_size, 3))
        b = np.random.uniform(-10,10,size=batch_size)
        v = np.sign(b) * np.random.uniform(0, 0.1, size=batch_size)
        x = np.random.uniform(b-v, b)
        M[:,0],M[:,1],M[:,2] = x,v,b
        yield M, (x+v).reshape(-1,1)
```

```python
i_b = Input(shape=(3,))

h_b = Dense(10, activation='linear')(i_b)

h_b = Dense(1, use_bias=False, activation='linear')(h_b)

out_b = h_b
# out = Dense(1, use_bias=False)(mul)
move = Model(inputs=i_b, outputs=out_b)
move.compile(loss=mean_squared_error,
              optimizer='adam',
              metrics=['accuracy'])
move.summary()
```

```python
move.fit_generator(move_gen(), steps_per_epoch=12800, epochs=10)
```

```python
# save_model(move, 'move')
move = open_and_load('move')
```

```python
X, Y = next(move_gen(batch_size=10000))
P = move.predict(X)
mse(Y, P) #2.442256002509147e-09
```

```python
plot_loss(move, move_gen)
```

```python
con = plot_landscape(move, move_gen)
```

```python
con = np.round(con, decimals=2)
plt.figure(figsize=(10,10))
cs = plt.contourf(con, levels=[0,0.05,0.1,0.15,0.2,0.25,0.3])
proxy = [plt.Rectangle((0,0),1,1,fc = pc.get_facecolor()[0]) 
    for pc in cs.collections]
plt.legend(proxy, ["0.00", "0.05","0.1","0.15","0.2","0.25"])
plt.savefig('loss_move.png', transparent=True)
plt.show()
```

### Bounce

```python
def bounce_gen(batch_size=128):
    while True:
        M = np.zeros((batch_size, 3))
        b = np.random.uniform(-10,10,size=batch_size)
        v = np.sign(b) * np.random.uniform(0, 0.1, size=batch_size)
        x = np.random.uniform(b-v, b)
        M[:,0],M[:,1],M[:,2] = x,v,b #-v not v probably
        yield M, (2*b-v-x).reshape(-1,1)
```

```python
i = Input(shape=(3,))

h = Dense(10, activation='linear')(i)
h = Dense(1, use_bias=False, activation='linear')(h)

out = h
# out = Dense(1, use_bias=False)(mul)
bounce = Model(inputs=i, outputs=out)
bounce.compile(loss=mean_squared_error,
              optimizer='adam',
              metrics=['accuracy'])
bounce.summary()
```

```python
bounce.fit_generator(bounce_gen(), steps_per_epoch=12800, epochs=10)
```

```python
# save_model(bounce, 'bounce')
bounce = open_and_load('bounce')
```

```python
X, Y = next(bounce_gen(batch_size=10000))
P = bounce.predict(X)
mse(Y, P) #2.8657695834505965e-08
```

```python
plot_loss(bounce, bounce_gen)
```

```python
con = plot_landscape(bounce, bounce_gen)
```

# Select move or bounce

```python
def att_gen(batch_size=128):
    while True:
        M = np.zeros((batch_size, 3))
        b = np.random.uniform(-10,10,size=batch_size)
        v = np.sign(b) * np.random.uniform(0, 0.1, size=batch_size)
        x = np.random.uniform(b-2*v, b)
        M[:,0],M[:,1],M[:,2] = x,v,b
        yield M, ((np.abs(x+v)-np.abs(b))>0 - 0.5).reshape(-1,1)
```

```python
i_att = Input(shape=(3,))

a1 = Dense(10, activation='relu')(i_att)
a1 = Dense(10, activation='relu')(a1)
a = Dense(1, activation='tanh')(a1)

a = change(a)

# out = Dense(1, use_bias=False)(mul)
att = Model(inputs=i_att, outputs=a)
att.compile(loss=mean_squared_error,
              optimizer='adam',
              metrics=['accuracy'])
att.summary()
```

```python
att.fit_generator(att_gen(), steps_per_epoch=12800, epochs=10)
```

```python
# save_model(att, 'att')
att = open_and_load('att')
```

```python
X, Y = next(att_gen(batch_size=1000))
P = att.predict(X)
mse(Y, P) #2.8657695834505965e-08
```

```python
plot_loss(att, att_gen)
```

# Combination

```python
i = Input(shape=(4,))

x = Lambda(lambda s: s[:,0],name='x')(i)
v = Lambda(lambda s: s[:,1],name='v')(i)
b1 = Lambda(lambda s: s[:,2],name='b1')(i)
b2 = Lambda(lambda s: s[:,3],name='b2')(i)
dropv = Lambda(lambda s: K.stack([s[0],s[1],s[2]], axis=1), output_shape=(3,),name='drop_v')([b1,b2,x])
# dropv = Concatenate(axis=1)([b1,b2])
# dropv = Reshape((3,))(dropv)
border.name='border'
getb = border(dropv)
getb = Reshape((1,))(getb)

xv = Lambda(lambda s: K.stack([s[0],s[1]], axis=1), output_shape=(2,),name='xv')([x,v])
# xv = Reshape((2,))(xv)
xvb = Concatenate(name='xvb')([xv,getb])
att.name='attention'
at = att(xvb)
# at3 = Lambda(lambda s: K.stack([s[:,0],s[:,0],s[:,0]],axis=1))(at)
bounce.name='bounce'
h3 = bounce(xvb)
v2 = Lambda(lambda s: -s, output_shape=(1,),name='new_v')(v)
h3 = Concatenate(name='bounce_concat')([h3,Reshape((1,))(v2)])
mul1 = Multiply(name='mul_att_bounce')([h3, at])

notat = Lambda(lambda s: 1-s, output_shape=(1,),name='1-att')(at)
# notat3 = Lambda(lambda s: K.stack([1-s[:,0],1-s[:,0],1-s[:,0]],axis=1))(at)
move.name='move'
g3 = move(xvb)
g3 = Concatenate(name='move_concat')([g3,Reshape((1,))(v)])
mul2 = Multiply(name='mul_att_move')([g3, notat])

out = Add(name='output')([mul1,mul2])

module = Model(inputs=i, outputs=out)
module.compile(loss=mean_squared_error,
              optimizer='adam',
              metrics=['accuracy'])
```

```python
module.summary()
```

```python
def gen(batch_size=128):
    while True:
        a = np.zeros((batch_size,4)) #x, v, b1, b2
        a[:,1] = np.random.uniform(0,0.1,size=batch_size)
        a[:,2] = np.random.uniform(1,-10,size=batch_size)
        a[:,3] = np.random.uniform(1, 10,size=batch_size)
        i = np.random.choice([0,1], size=batch_size) #isb1
        b = a[:,2]*i + a[:,3]*(1-i)
        a[:,1] *= np.sign(1-i-0.5) 
        a[:,0] = np.random.uniform(b, b-10*a[:,1])
        
        isBounce = ( np.abs(a[:,0] + a[:,1]) - np.abs(b)) > 0
        
        v = a[:,1] * np.sign(1-isBounce-0.5) #new speed
        x = (a[:,0] + a[:,1]) * (1-isBounce) + (2*b-a[:,1]-a[:,0])*isBounce
        b_newIsb1 = ((x-a[:,2])/(a[:,3]-a[:,2])) < 0.5
        b_new = a[:,2]*b_newIsb1 + a[:,3]*(1-b_newIsb1)
        y = np.zeros((batch_size,2))
        y[:,0] = x
        y[:,1] = v
        yield a, y     
```

```python
X, Y = next(gen(batch_size=1))
P = module.predict(X)
```
